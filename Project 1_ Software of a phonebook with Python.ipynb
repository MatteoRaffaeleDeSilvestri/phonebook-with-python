# Add new contacts

We now enrich the various functions of the Phonebook class.

Let’s start by defining the **add_contact** functionality that allows a new contact to be added to the address book. This feature defines the structure of the information contained in the address book. In particular, we will save the following information for your contact:

* Name;
* Surname;
* Address;
* Primary telephone number;
* Secondary telephone number;
* Primary email;
* Secondary email.

Since the general structure of the heading is a dictionary, to create a new contact we will define as key value the concatenation of name and surname and as value an additional dictionary containing the remaining information.

In this phase we will also mplement a way to detect and manage if the user is trying to add a contact that is alredy in the phonebook.

```python
class Phonebook:

    """
    Initialization of the class phonebook
    """

    def __init__(self):

        """
        Definition of the empty phonebook
        """

        self.pb = dict() # Empty phonebook

    def add_contact(self, contact_info):

        """
        Add new contact to the phonebook
        """

        name = contact_info['name']
        surname = contact_info['surname']
        address = contact_info['address']
        phone_number1 = contact_info['phone_number_1']
        phone_number2 = contact_info['phone_number_2']
        email1 = contact_info['email1']
        email2 = contact_info['email2']

        # Check for duplicate contacts
        if (name, surname) in self.pb.keys():
            replace_contact = input(f'A contact for {name} {surname} already exist. Do you want to replace it? [Y/N] ')

            while replace_contact.upper() not in ('Y','N'):
                replace_contact = input(f'Please select one of the options. A contact for {name} {surname} already exist. Do you want to replace it? [Y/N] ')

            if replace_contact.upper() == 'Y':
                self.pb[(name, surname)] = {'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2}
                print(f'Contact for {name} {surname} replaced successfully.')
            else:
                print(f'Contact for {name} {surname} not replaced.')

        else:
            self.pb[(name, surname)] = {'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2}
            print(f'Contact for {name} {surname} added successfully.')
```

```python
# @markdown <h1><b>Phonebook</b></h1> { display-mode: "form" }

# @markdown <h6><i>by ContactEase</i></h6>

# @markdown Run this cell to initialise the phonebook.</br>
# @markdown This section also offer the opportunity to insert some dummy contacts to the phonebook wich may be useful to run some tests.

import pandas as pd

class Phonebook:

    """
    Initialization of the class phonebook
    """

    def __init__(self):

        """
        Definition of the empty phonebook
        """

        self.pb = dict() # Empty phonebook

    def add_contact(self, contact_info):

        """
        Add new contact to the phonebook
        """

        name = contact_info['name']
        surname = contact_info['surname']
        address = contact_info['address']
        phone_number1 = contact_info['phone_number_1']
        phone_number2 = contact_info['phone_number_2']
        email1 = contact_info['email1']
        email2 = contact_info['email2']

        # Check for duplicate contacts
        if (name, surname) in self.pb.keys():
            replace_contact = input(f'A contact for {name} {surname} already exist. Do you want to replace it? [Y/N] ')

            while replace_contact.upper() not in ('Y','N'):
                replace_contact = input(f'Please select one of the options. A contact for {name} {surname} already exist. Do you want to replace it? [Y/N] ')

            if replace_contact.upper() == 'Y':
                self.pb[(name, surname)] = {'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2}
                print(f'Contact for {name} {surname} replaced successfully.')
            else:
                print(f'Contact for {name} {surname} not replaced.')

        else:
            self.pb[(name, surname)] = {'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2}
            print(f'Contact for {name} {surname} added successfully.')

    def view_contacts(self):

        """
        View all the contacts in the phonebook
        """

        pb_ls = list()

        for item in my_phonebook.pb.items():
            pb_ls.append(list(item[0]) + list(item[1].values()))

        # Print the phonebook
        df = pd.DataFrame(pb_ls, columns=['Name','Surname','Address','Primary phone number','Secondary phone number','Primary email','Secondary email'])
        print(df.to_markdown(index=False))

    def edit_contact(self, contact_to_edit):

        """
        Edit a contact from the phonebook
        """

        name = contact_to_edit['name']
        surname = contact_to_edit['surname']
        address = contact_to_edit['address']
        phone_number1 = contact_to_edit['phone_number_1']
        phone_number2 = contact_to_edit['phone_number_2']
        email1 = contact_to_edit['email1']
        email2 = contact_to_edit['email2']

        # Check for the existing contact
        if (name, surname) not in self.pb.keys():
            insert_contact = input(f'A contact for {name} {surname} does not exist. Do you want to create a new contact? [Y/N] ')

            while insert_contact.upper() not in ('Y','N'):
                insert_contact = input(f'Please select one of the options. A contact for {name} {surname} does not exist. Do you want to create a new contact? [Y/N] ')

            if insert_contact.upper() == 'Y':
                self.add_contact({'name':name,'surname':surname,'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2})
            else:
                print(f'Contact for {name} {surname} not created.')

        else:
            self.pb[(name, surname)] = {'address':address,'phone_number_1':phone_number1,'phone_number_2':phone_number2,'email1':email1,'email2':email2}
            print(f'Contact for {name} {surname} modified successfully.')

    def delete_contact(self, contact_to_delete):

        """
        Delete a contact from the phonebook
        """

        name = contact_to_delete['name']
        surname = contact_to_delete['surname']

        # Check for the existing contact
        if (name, surname) not in self.pb.keys():
            print(f'A contact for {name} {surname} does not exist.')
        else:
            del self.pb[(name, surname)]
            print(f'Contact for {name} {surname} deleted successfully.')

    def search_contacts(self, contact_to_search):

        """
        Search a contact in the phonebook
        """

        search_by = contact_to_search['by']
        value = contact_to_search['val']

        pb_ls = list()

        for item in my_phonebook.pb.items():

            name = item[0][0]
            surname = item[0][1]
            address = list(item[1].values())[0]
            primary_phone_number = list(item[1].values())[1]
            secondary_phone_number = list(item[1].values())[2]
            primary_email = list(item[1].values())[3]
            secondary_email = list(item[1].values())[4]

            if search_by == 'name' and name == value:
                pb_ls.append([name, surname, address, primary_phone_number, secondary_phone_number, primary_email, secondary_email])
            elif search_by == 'surname' and surname == value:
                pb_ls.append([name, surname, address, primary_phone_number, secondary_phone_number, primary_email, secondary_email])
            elif search_by == 'address' and address == value:
                pb_ls.append([name, surname, address, primary_phone_number, secondary_phone_number, primary_email, secondary_email])
            elif search_by == 'phone' and value in (primary_phone_number, secondary_phone_number):
                pb_ls.append([name, surname, address, primary_phone_number, secondary_phone_number, primary_email, secondary_email])
            elif search_by == 'mail' and value in (primary_email, secondary_email):
                pb_ls.append([name, surname, address, primary_phone_number, secondary_phone_number, primary_email, secondary_email])

        # Print the phonebook
        fd = pd.DataFrame(pb_ls, columns=['Name','Surname','Address','Primary phone number','Secondary phone number','Primary email','Secondary email'])
        if not(fd.size):
            print(f'No contacts have been found for those informations.')
        else:
            df = pd.DataFrame(pb_ls, columns=['Name','Surname','Address','Primary phone number','Secondary phone number','Primary email','Secondary email'])
            print(df.to_markdown(index=False))

    def import_contacts(self, path_to_csv_file):

        """
        Import data from csv file
        """

        # Set/Reset phonebook
        self.pb = dict()

        try:
            import_pb = pd.read_csv(path_to_csv_file, keep_default_na=False)

        # Manage empty csv file
        except pd.errors.EmptyDataError:
            print('There are no contacts to import.')

        else:
            print('Phonebook imported successfully.')
            import_pb = pd.read_csv(path_to_csv_file, keep_default_na=False)

            for row in import_pb.values.tolist():
                self.pb[(row[0],row[1])] = {'address':row[2],'phone_number_1':row[3],'phone_number_2':row[4],'email1':row[5],'email2':row[6]}

            # Print the phonebook
            self.view_contacts()

    def export_contacts(self, path_to_csv_file):

        """
        Save the current phonebook in a csv file
        """

        pb_ls = list()

        for item in my_phonebook.pb.items():
            pb_ls.append(list(item[0]) + list(item[1].values()))

        print('Phonebook exported successfully.')

        # Prepare for export
        export_pb = pd.DataFrame(pb_ls, columns=['Name','Surname','Address','Primary phone number','Secondary phone number','Primary email','Secondary email'])
        export_pb.to_csv(path_to_csv_file, index=False)

# Initialise phonebook
my_phonebook = Phonebook()

# Setup default phonebook
gen_pb = input(f'Phonebook initialised sucessfully.\nDo you also want to add some test contact to it? [Y/N] ')

while gen_pb.upper() not in ('Y','N'):
    gen_pb = input(f'Please select one of the options. Do you also want to add some test contact to it? [Y/N] ')

if gen_pb.upper() == 'Y':

    # Add contacts
    my_phonebook.add_contact({'name':'Vernon', 'surname':'Hartman','address':'2038 Medical Center Drive - Plant City (FL)','phone_number_1':'2011847498','phone_number_2':'2017535617','email1':'meinauppiddulle-5178@yopmail.com','email2':''})
    my_phonebook.add_contact({'name':'Vivian', 'surname':'Robbins','address':'','phone_number_1':'2014003144','phone_number_2':'','email1':'','email2':''})
    my_phonebook.add_contact({'name':'Billy', 'surname':'Kelley','address':'1309 Hemlock Lane - Harlingen (TX)','phone_number_1':'2018160148','phone_number_2':'','email1':'ruwiwassauru-9554@yopmail.com','email2':''})
    my_phonebook.add_contact({'name':'Frank', 'surname':'Case','address':'1821 Fulton Street - Parkersburg (WV)','phone_number_1':'2015360223','phone_number_2':'','email1':'','email2':''})
    my_phonebook.add_contact({'name':'Chris', 'surname':'Hanson','address':'','phone_number_1':'2013416439','phone_number_2':'2012468131','email1':'','email2':''})
    my_phonebook.add_contact({'name':'Dana', 'surname':'Floyd','address':'4525 Hardman Road - Springfield (VT)','phone_number_1':'2019852828','phone_number_2':'','email1':'tekeddeizeureu-4376@yopmail.com','email2':''})
    my_phonebook.add_contact({'name':'Robert', 'surname':'Landry','address':'1213 Pinnickinick Street - Bellingham (WA)','phone_number_1':'2016779026','phone_number_2':'2011095631','email1':'brivajetemma-2078@yopmail.com','email2':'brivajetemma-2079@yopmail.com'})
    my_phonebook.add_contact({'name':'Jordan', 'surname':'Cole','address':'','phone_number_1':'2016682811','phone_number_2':'','email1':'wujokeuyicri-6592@yopmail.com','email2':''})
    my_phonebook.add_contact({'name':'Eduard', 'surname':'Russell','address':'','phone_number_1':'2011948516','phone_number_2':'2018032328','email1':'haunnaffauf@yopmail.com','email2':'haunnaffauffaddau@yopmail.com'})
    my_phonebook.add_contact({'name':'Cassidy', 'surname':'Adkins','address':'418 Despard Street - East Point (GA)','phone_number_1':'2017392642','phone_number_2':'','email1':'wucrennauprouja-9911@yopmail.com','email2':''})
    print(f'Test phonebook generated successfully.')

else:
    print(f'Test phonebook not generated.')

```
